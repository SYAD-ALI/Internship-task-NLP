##Task 1: Conceptual Understanding

#1. What is the difference between "Love" and "love" in NLP?

In NLP, "Love" and "love" are treated as two different words because systems are case-sensitive.
This increases the number of unique words (vocabulary) and can affect model performance.
So, we usually convert all text to lowercase during preprocessing to treat them as the same word.

2. What happens if stopwords are not removed?

If stopwords are not removed, the dataset becomes noisy.
Common words like "is", "the", "and" take more space and reduce the importance of meaningful words.
This can slow down model training and increase memory usage.

3. Two real-world scenarios where removing stopwords can be harmful:

In sentiment analysis, words like "not", "no", "never" are very important.
For example, "I am happy" and "I am not happy" have opposite meanings.

In question answering systems, words like "what", "where", "how" are important to understand the question properly.

4. Difference between Stemming and Lemmatization:

Both stemming and lemmatization are used to convert words into their base form.

Stemming is a simple method that cuts words (example: "running" → "runn"), which may not always be correct.

Lemmatization is more accurate because it converts words into meaningful base forms using language rules
(example: "running" → "run").

#1. Stemming

Stemming is a simple process used in NLP where words are reduced to their root form by removing prefixes or suffixes.

- It follows basic rules (not very smart 😄)
- It does not understand the meaning or context of the word
- Sometimes it gives incorrect or incomplete words

Examples:
- "running" → "run"
- "studies" → "studi"
- "happiness" → "happi"

Advantages:
- Very fast
- Works well for large datasets

Disadvantages:
- Less accurate
- Output may not be real words


2. Lemmatization

Lemmatization is a more advanced method that converts words into their correct base or dictionary form.

- It uses vocabulary and language rules
- It understands context and part of speech
- Always gives meaningful words

Examples:
- "running" → "run"
- "better" → "good"
- "studies" → "study"

Advantages:
- More accurate than stemming
- Produces valid and meaningful words

Disadvantages:
- Slower than stemming
- Needs more processing power


Key Differences:

- Stemming is rule-based, while lemmatization uses dictionary and context
- Stemming is faster, lemmatization is slower but more accurate
- Stemming may give wrong words, lemmatization gives correct words
- Stemming does not understand context, lemmatization does


Conclusion:

Stemming is useful when speed is important and we have large data.
Lemmatization is better when we need accuracy and meaningful results,
especially in applications like chatbots, search engines, and sentiment analysis.

##Task 2 Preprocessing Function

```python
#Task 2 Preprocessing Function

import re
import string
from collections import Counter

def preprocess_text(text):

    # Error handling
    if not isinstance(text, str):
        return [], ""

    text = text.strip()

    if text == "":
        return [], ""

    # Remove URLs
    text = re.sub(r"http\S+|www\S+", "", text)

    # Remove emails
    text = re.sub(r"\S+@\S+", "", text)

    # Remove numbers
    text = re.sub(r"\d+", "", text)

    # Convert to lowercase
    text = text.lower()

    # Handle repeated characters
    text = re.sub(r"(.)\1{2,}", r"\1", text)

    # Remove punctuation
    text = text.translate(str.maketrans("", "", string.punctuation))

    # Remove emojis / special characters
    text = re.sub(r"[^\w\s]", "", text)

    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()

    # Tokenization
    tokens = text.split()

    # Remove short tokens except "no", "not"
    tokens = [t for t in tokens if len(t) > 2 or t in ["no", "not"]]

    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence

    # Test input
text = "Soooo goooood!!! 😍😍"

# Call function
tokens, cleaned = preprocess_text(text)

# Print output
print("Tokens:", tokens)
print("Cleaned:", cleaned)
```
